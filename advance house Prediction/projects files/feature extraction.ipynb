```python
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
pd.set_option('display.max_columns',None)

#pd.set_option('display.max_rows',None)
```

```python
df=pd.read_csv('train.csv')
df
```

```python
feature_nan=[feature for feature in df.columns if df[feature].isnull().sum()>1 if df[feature].dtypes=='O']
for feature in feature_nan:
    print(feature,'{}'.format(df[feature].isnull().sum()))
```

```python
for feature in feature_nan:
    df[feature].fillna('missing',inplace=True)
```

handeling missing vals

```python
num_val_nan=[feature for feature in df.columns  if df[feature].isnull().sum()>=1  if df[feature].dtype !='O']
num_val_nan
```

```python
for feature in num_val_nan:
    median_val=df[feature].median()
    df[feature+'nan']=np.where(df[feature].isnull(),1,0)
    df[feature].fillna(median_val, inplace=True)
```

```python
for feature in num_val_nan:
    print(feature,'{}'.format(df[feature].isnull().sum()))
df.head()
```

```python
## year month 


for feature in ['YearBuilt', 'YearRemodAdd', 'GarageYrBlt']:
    df[feature]=df['YrSold']-df[feature]
    
df[['YearBuilt', 'YearRemodAdd', 'GarageYrBlt']].head()
```

```python
num_features=['LotFrontage' ,'LotArea','GrLivArea','1stFlrSF']
for feature in num_features:
    df[feature]=np.log(df[feature])
```

```python
df.head()
```

```python
cat_vars=[feature for feature in df.columns if df[feature].dtypes=='O']
cat_vars
```

```python
for feature in cat_vars:
    temp=df.groupby(feature)['SalePrice'].count()/len(df)
    temp_df=temp[temp>0.01].index
    df[feature]=np.where(df[feature].isin(temp_df),df[feature],'Rare_var')
df
```

## feature scaling

```python
from sklearn.preprocessing import MinMaxScaler , LabelEncoder
from sklearn import preprocessing
```

```python
scaler=MinMaxScaler()
label_enco=LabelEncoder()
```

```python
for feature in cat_vars:
    df[feature]=label_enco.fit_transform(df[feature])
```

```python
df
```

```python
feature_scaler=[feature for feature in df.columns if feature  not in ['Id','SalePrice'] ]
scaler.fit(df[feature_scaler])
```

```python
data=pd.concat([df[['Id','SalePrice']].reset_index(drop=True),
                 pd.DataFrame(scaler.transform(df[feature_scaler]),columns=feature_scaler)],axis=1)
```

```python
data.to_csv('x_train.csv',index=False)
```
