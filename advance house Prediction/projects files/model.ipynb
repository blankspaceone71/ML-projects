```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn.linear_model import Lasso
from sklearn.feature_selection import SelectFromModel
#print(M)
```

```python
df=pd.read_csv('x_train.csv')
#df['SalePrice']=np.log(df['SalePrice'])
#df.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
df
```

```python
x_train=df.drop(['Id','SalePrice'],axis=1)
d={}
for i in x_train.columns:
    d[i]=x_train[i].isnull().sum()
    #(x_train[i].isnull().sum(),i)
d    
zx=[]
for d,v in d.items():
    if v>1:
        zx.append(d)
zx
for f in zx:
    median_val=x_train[f].median()
    x_train[f].fillna(median_val,inplace=True)
```

```python

d={}
for i in x_train.columns:
    d[i]=x_train[i].isnull().sum()
d
```

```python
y_train=df.SalePrice
```

```python
feat_select_model=SelectFromModel(Lasso(alpha=0.005,normalize=True, max_iter=1e5,random_state=0))
feat_select_model.fit(x_train,y_train) ## lasso 
```

```python
feat_select_model.get_support()## getting important features(columns) in the dataset 
```

```python
selected_columns=x_train.columns[(feat_select_model.get_support())]
selected_columns
```

```python
print('total features :{}'.format(x_train.shape[1]))
print('selected featues :{}'.format(len(selected_columns)))
print('features whose co efficient is zero: {}'.format(np.sum(feat_select_model.estimator_.coef_==0)))
```
