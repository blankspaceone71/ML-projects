```python
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
#pd.set_option('display.max_rows',None)
```

```python
df=pd.read_csv('test.csv')
len(df.columns)
```

```python
cat_feat_nan=[feature for feature in df.columns  if df[feature].isnull().sum()>=1 if df[feature].dtypes=='O']
for feature in cat_feat_nan:
    print(feature,'{}'.format(df[feature].isnull().sum()))
```

```python
for feature in cat_feat_nan:
    df[feature].fillna('missing',inplace=True)
```

```python
for feature in cat_feat_nan:
    print(feature,'{}'.format(df[feature].isnull().sum()))
```

```python
num_feat_nan=[feature for feature in df.columns if df[feature].isnull().sum()>1 if df[feature].dtypes !='O']
num_feat_nan
```

```python
num_ft=['LotFrontage', 'MasVnrArea','GarageYrBlt']
```

```python
for feature in num_ft:
    mideann=df[feature].median()
    df[feature+'nan']=np.where(df[feature].isnull(),1,0)
    df[feature].fillna(mideann,inplace=True)
    
```

```python
df
```

```python
from sklearn.preprocessing import MinMaxScaler , LabelEncoder
from sklearn import preprocessing
```

```python
scaler=MinMaxScaler()
label_enco=LabelEncoder()
cat_vars=[feature for feature in df.columns if df[feature].dtypes=='O']
```

```python
for i in cat_vars:
    df[i]=label_enco.fit_transform(df[i])
df
```

```python
feature_scaler=[feature for feature in df.columns if feature  not in ['Id'] ]
scaler.fit(df[feature_scaler])
```

```python
data=pd.concat([df['Id'].reset_index(drop=True),
                 pd.DataFrame(scaler.transform(df[feature_scaler]),columns=feature_scaler)],axis=1)
```

```python
data.to_csv('scaled_test_data.csv',index=False)
```
